#### Probamos otras alternativas de ver la cantidad de revisiones

```python
df_log_revisiones = df.copy()
df_log_revisiones['log1p'] = np.log1p(df_log_revisiones['cantidad_revisiones'])
df_log_revisiones['log2'] = np.log2(df_log_revisiones['cantidad_revisiones'])
df_log_revisiones['log10'] = np.log10(df_log_revisiones['cantidad_revisiones'])
df_log_revisiones['log'] = np.log(df_log_revisiones['cantidad_revisiones'])

plt.rcParams["figure.figsize"] = (14,16)
f, axes = plt.subplots(4, 1)
f.tight_layout(pad=5.0)

g1 = sns.boxplot(data=df_log_revisiones, y='log1p', x='disponibilidad', orient='v', ax=axes[0])
g1.set(title='log1p')

g2 = sns.boxplot(data=df_log_revisiones, y='log2', x='disponibilidad', orient='v', ax=axes[1])
g2.set(title='log2')

g3 = sns.boxplot(data=df_log_revisiones, y='log10', x='disponibilidad', orient='v', ax=axes[2])
g3.set(title='log10')

g4 = sns.boxplot(data=df_log_revisiones, y='log', x='disponibilidad', orient='v', ax=axes[3])
g4.set(title='log')

plt.show()
```

```python
plt.rcParams["figure.figsize"] = (14,8)
f, axes = plt.subplots(2, 1)
f.tight_layout(pad=5.0)

g1 = sns.boxplot(data=df_log_revisiones, y='cantidad_revisiones', x='disponibilidad', orient='v', ax=axes[0])
g1.set(title='Cantidad de revisiones')

g4 = sns.boxplot(data=df_log_revisiones, y='log', x='disponibilidad', orient='v', ax=axes[1])
g4.set(title='log Cantidad de revisiones')

plt.show()
```

```python
q = 0.95

df_umbrales_por_disponibilidad = df_log_revisiones.groupby(['disponibilidad'],as_index=False).agg(
        umbral_log_revisiones_minimo = ('log', lambda x: x.quantile(q))
)

df_umbrales_por_disponibilidad
```

```python
file_umbrales_por_disponibilidad = 'dataset/umbrales_por_disponibilidad.csv'
df_umbrales_por_disponibilidad.to_csv(file_umbrales_por_disponibilidad, index=False)
```

#### Asignamos la clase al dataset principal

```python
df_cervezas_unicas_resumen_valores_con_clase = pd.merge(df, df_umbrales_por_disponibilidad, on='disponibilidad')
df_cervezas_unicas_resumen_valores_con_clase.head()


df_cervezas_unicas_resumen_valores_con_clase['es_popular'] = np.where(
    np.log(df_cervezas_unicas_resumen_valores_con_clase['cantidad_revisiones'])>df_cervezas_unicas_resumen_valores_con_clase['umbral_log_revisiones_minimo'], 'SI', 'NO')

df_cervezas_unicas_resumen_valores_con_clase.head()
```

```python
file_cervezas_unicas_resumen_valores_con_clase = 'dataset/cervezas_unicas_resumen_valores_con_clase.csv'
df_cervezas_unicas_resumen_valores_con_clase.to_csv(file_cervezas_unicas_resumen_valores_con_clase, index=False)
```
